```python
"""Zadanie 5 – Wykrywanie NaN
Sprawdź które kolumny w Titanicu mają brakujące dane.
Wymagania:
- Policz NaN w każdej kolumnie
- Oblicz procent NaN
- Wyświetl tylko kolumny z >10% braków
"""

import pandas as pd

# 1. Define the dataset URL: 
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# 2. Load the data directly from the URL info a DataFrame:
titanic = pd.read_csv(url)

print("Number of NaN in every column:")
print(titanic.isnull().sum())

missing_percent = (titanic.isnull().sum()/len(titanic) * 100).round(2)

print("\nColumns with more than 10% missing values")
print(missing_percent[missing_percent > 10])
```

```python
"""Zadanie 6 – Wypełnianie NaN średnią
Wypełnij brakujące wartości w kolumnie 'Age' średnią.
Wymagania:
- Oblicz średnią wieku (bez NaN)
- Wypełnij NaN tą średnią
- Sprawdź czy nie ma już NaN w 'Age'
"""

import pandas as pd 

# 1. Load the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# 2. Create a working copy
titanic = pd.read_csv(url)

# 3. Calculate the mean age (excluding NaNs)
titanic_filled = titanic.copy()

# 3. Calculate the mean age (excluding NaNs)
age_mean = titanic['Age'].mean()

# 4. Fill missing values with the mean using the modern assignment method (no inplace=True)
titanic_filled['Age'] = titanic_filled['Age'].fillna(age_mean)

print(f"\nAge column statistics after filling NaN:")
print(f"Used average age:{age_mean:.2f}")
print(titanic_filled['Age'].describe())

#5. Verify that there are no Non values left in the 'Age' column
remaining_nans = titanic_filled['Age'].isnull().sum()
print(f"\nVerification: Number of missing values in 'Age' column: {remaining_nans}" )
```

```python
"""Zadanie 7 – Sortowanie DataFrame
Posortuj pasażerów Titanica po wieku (od najmłodszego do najstarszego).
Wymagania:
- Użyj .sort_values()
- Wyświetl 10 najmłodszych
- Wyświetl 10 najstarszych
"""

import pandas as pd

# 1. Load the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# 2. Sort the DataFrame by 'Age' in ascending order (youngest to oldest)
titanic_sorted = titanic.sort_values('Age')

#3. Display the 10 youngest passengers (first 10 rows of the sorted DataFrame)
print("Top 10 Youngest Passengers")
#Selecting only 'Name' and 'Age' columns for better readability
print(titanic_sorted[['Name', 'Age']].head(10))

# 4. Display the 10 oldest passengers (last 10 rows of the sorted DataFrame)
titanic_oldest = titanic.sort_values('Age', ascending=False)
print("\n Top 10 Oldest Passengers")
print(titanic_oldest[['Name', 'Age']].head(10))
```

```python
"""Zadanie 8 – Filtrowanie przez .query()
Użyj metody .query() aby znaleźć pasażerów 1. klasy, którzy zapłacili więcej niż 50£.
Wymagania:
Użyj .query() z warunkiem złożonym
Policz ile było takich pasażerów
Oblicz średnią cenę ich biletów
"""

import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# 2. Use .query() to find 1st class passengers who paid more than 50
# Logic: Pclass 1 is represented by integer 1, Fare is a float
rich_passengers = titanic.query('Pclass == 1 and Fare > 50')

# 3. Count the number of such passangers:
count = len(rich_passengers)

# 4. Calculate the average fare for this specific group
average_fare = rich_passengers['Fare'].mean()

# 5. Display the results in a professional report format
print("1st Class/ Fare > 50 Analysis")
print(f"Number of passengers matching criteria: {count}")
print(f"Average ticket price in this group: {average_fare:.2f} GBP")
```

```python
"""Zadanie 9 – Batch processing
Wczytaj dataset Titanic w batchach po 100 wierszy i oblicz średni wiek w każdym batchu.
Wymagania:
Użyj chunksize=100
Dla każdego batcha wypisz numer i średni wiek
Oblicz globalną średnią ze wszystkich batchy
"""

import pandas as pd

# 1. Define the dataset URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

total_rows = 0
age_sum = 0
age_count = 0
batch_number = 1

print("Batch Processing Report")

# 2. Process data in chunks of 100 rows
for chunk in pd.read_csv(url, chunksize = 100):

    #Calculate average age for the current batch
    batch_mean = chunk['Age'].mean()
    print(f"Batch {batch_number}: Processed {len(chunk)} rows. Average age: {batch_mean:.2f}")

    #Accumulate global stats for final calculation
    total_rows += len(chunk)
    age_sum += chunk['Age'].sum()
    age_count += chunk['Age'].count()

    #Increment batch counter
    batch_number += 1

# 3. Calculate and display global statistics
global_mean = age_sum / age_count

print("\n Global Summary")
print(f"Total rows processed: {total_rows} wierszy")
print(f"Global average age: {global_mean:.2f}")
```

```python
"""Zadanie 10 – Kompleksowe czyszczenie danych
Wyczyść dataset Titanic stosując pełny pipeline:
- Usuń kolumny z >60% NaN
- Wypełnij Age medianą
- Wypełnij Embarked dominantą
- Usuń wiersze z pozostałymi NaN
Wymagania:
- Przed i po: wypisz kształt DataFrame
- Sprawdź czy nie ma NaN
- Porównaj rozkład Age przed i po
"""

import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print(f"Original shape: {df.shape}")

limit = len(df) * 0.4
df_cleaned = df.dropna(thresh=limit, axis=1).copy()

age_median = df_cleaned['Age'].median()
df_cleaned['Age'] = df_cleaned['Age'].fillna(age_median)

embarked_mode = df_cleaned['Embarked'].mode()[0]
df_cleaned['Embarked'] = df_cleaned['Embarked'].fillna(embarked_mode)

df_cleaned = df_cleaned.dropna()

print("--- Final Report ---")
print(f"Cleaned shape: {df_cleaned.shape}")
print(f"Remaining NaNs: {df_cleaned.isnull().sum().sum()}")
```

```python
"""Zadanie 11 – Feature engineering
Stwórz nowe zmienne z Titanic:
- FamilySize = SibSp + Parch + 1
- IsAlone = 1 jeśli FamilySize == 1, inaczej 0
- AgeGroup = kategorie (0-12, 13-18, 19-35, 36-60, 60+)
Wymagania:
- Dodaj wszystkie 3 zmienne do DataFrame
- Wyświetl value_counts() dla każdej
- Oblicz wskaźnik przeżycia dla IsAlone vs nie alone
"""

import pandas as pd

df_cleaned['FamilySize'] = df_cleaned['SibSp'] + df_cleaned['Parch'] + 1

df_cleaned['IsAlone'] = (df_cleaned['FamilySize'] == 1).astype(int)

bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']
df_cleaned['AgeGroup'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels)

survival_rate = df_cleaned.groupby('IsAlone')['Survived'].mean()

print("--- Feature Engineering Results ---")
print(df_cleaned[['FamilySize', 'IsAlone', 'AgeGroup']].head())
print(f"\nSurvival Rate (0=Not Alone, 1=Alone):\n{survival_rate}")
```
