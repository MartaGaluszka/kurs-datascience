```python
"""Przykład 2: Transformacja danych (map + lambda)
Cel: Przeliczenie waluty.

Funkcja map(funkcja, kolekcja) bierze każdy element z listy i modyfikuje
 go według przepisu z lambdy.
(Ważna uwaga architektoniczna: w Pythonie wynik map trzeba "otoczyć" funkcją list(), 
aby z powrotem stał się czytelną listą, np.: list(map(...))).

Polecenie: Użyj map() i lambda, aby stworzyć nową listę ceny_eur. Lambda powinna brać 
każdą cenę z listy ceny_pln i dzielić ją przez 4.3 (przyjmijmy taki kurs euro). 
"""

# Lista samych cen (w PLN)
ceny_pln = [150, 45, 320, 89, 500, 20]

ceny_eur = list(map(lambda x: round(x / 4.3, 2), ceny_pln))

print(ceny_eur)
```

```python
"""Przykład 4: Sortowanie z kluczem (sorted + lambda)
Cel: Uporządkowanie złożonych struktur.

Funkcja sorted(kolekcja, key=lambda ...) pozwala sortować
dane nie tylko alfabetycznie, ale według konkretnego "klucza".

Polecenie: Mamy listę magazyn, gdzie każdy element to 
krotka (Nazwa, Ilość). Element na pozycji 0 to nazwa, a na 
pozycji 1 to ilość. Użyj sorted() z kluczem key=lambda x: ..., 
aby posortować magazyn według ilości sztuk (od najmniejszej do
największej). Wypisz wynik na ekran. 
"""
# Lista produktów w magazynie zapisana jako krotki: (Nazwa, Ilość sztuk)
magazyn = [("Laptop", 5), ("Myszka", 120), ("Klawiatura", 35), ("Monitor", 8)]

posortowany_magazyn = sorted(magazyn, key=lambda x: x[1])

print(posortowany_magazyn)
```

```python
"""Zadanie 6 – Funkcja z wartościami domyślnymi
Napisz funkcję powitanie(imie, jezyk='polski') , która wyświetla powitanie w
wybranym języku (polski, angielski, niemiecki).
Dataset:
Brak (funkcja operuje na stringach)
Wymagania:
Domyślny język to polski
Obsługa 3 języków
Wywołaj funkcję 3 razy z różnymi parametrami

Oczekiwany rezultat:
powitanie('Anna') → 'Cześć, Anna!'
powitanie('Anna', 'angielski') → 'Hello, Anna!'"""

def powitanie(imie, jezyk='polski'):
    if jezyk == 'polski':
        return f"Cześć, {imie}!"
    
    elif jezyk == 'angielski':
        return f"Hello, {imie}!"
    
    elif jezyk == 'niemiecki':
        return f"Hallo, {imie}!"   
    
print(powitanie('Anna'))
print(powitanie('Anna', 'angielski'))
print(powitanie('Tomasz', 'niemiecki'))            
```

```python
"""Zadanie 8 – Map i filter
Używając map() i filter() z lambda, przetwórz listę liczb: najpierw zostaw tylko
dodatnie, potem podwój każdą.
Dataset:
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

Wymagania:
filter() z lambda: x > 0
map() z lambda: x * 2
Konwertuj wynik na listę

Oczekiwany rezultat:
Lista: [6, 16, 14, 8, 12] (tylko dodatnie i podwojone)"""

liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

#Podejscie 1:
l_dodatnie = list(filter(lambda l: l > 0, liczby))
l_dodatnie_podwojone = list(map(lambda p: p * 2, l_dodatnie))

print(f" Wyniki z załozen z treści zadania:")
print(f"Liczby dodatnie = {l_dodatnie}")
print(f"Liczby dodatnie podwojone = {l_dodatnie_podwojone}")

#Podejście 2:
dodatnie_podwojne = [l * 2 for l in liczby if l > 0]

print(f"\n Wyniki z obliczeń drugim sposobem:")
print(f"Liczby dodatnie podwojne (II sposob) = {dodatnie_podwojne}")
```

```python
"""Zadanie 9 – Funkcja normalizacji min-max
Napisz funkcję normalize_minmax(dane) , która normalizuje listę liczb 
do zakresu [0, 1] używając metody min-max.
Dataset:
wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]

Wymagania:
Wzór: (x - min) / (max - min)
Funkcja zwraca nową listę znormalizowanych wartości
Zaokrąglij do 3 miejsc po przecinku

Oczekiwany rezultat:
Lista wartości w zakresie [0.0, 1.0]
"""

def normalize_minmax(dane, metoda="min-max", zaokraglenie=3):
    """
    Normalizuje dane przy użyciu wybranej metody.

    Args:
        dane (list): Lista wartości do normalizacji
        metoda (str): Metoda normalizacji ("min-max"")
        zaokraglenie (int): Liczba miejsc po przecinku

    Returns:
        list: Znormalizowane dane
    """

    if metoda == "min-max":
        #Normalizacja min-max: (x - min) / (max - min)
        min_val = min(dane)
        max_val = max(dane)
        zakres = max_val - min_val

        znormalizowane = []
        for wartosc in dane:
            norma = (wartosc - min_val) / zakres
            znormalizowane.append(round(norma, zaokraglenie))
        return znormalizowane
        
# Dane:
wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]        

#Normalizacja min-max (domyślna)
wyn_norm_minmax = normalize_minmax(wynagrodzenia)
print(f"Oryginalne wynagrodzenia: {wynagrodzenia}")
print(f"Normalizacja min_max: {wyn_norm_minmax}")
```

```python
'''
Zadanie 10 – Dictionary comprehension - agregacja
Masz listę transakcji: [{"produkt": "X", "wartosc": 100}, ...] . Używając dict
comprehension i grupowania, oblicz łączną wartość dla każdego produktu.
Dataset:
transakcje = [
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "klawiatura", "wartosc": 249},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "mysz", "wartosc": 89}
]

Wymagania:
Pogrupuj po produkcie
Zsumuj wartości dla każdego produktu
Wynik jako słownik: {"laptop": 5998, ...}

Oczekiwany rezultat:
Słownik z sumą wartości per produkt
'''

#Dataset:
transakcje = [
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "klawiatura", "wartosc": 249},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "mysz", "wartosc": 89}
]

#Grupowanie po produktach:
from collections import defaultdict
grouped = defaultdict(list)

for t in transakcje:
    grouped[t["produkt"]].append(t["wartosc"])

#Dictionary comprehension - obliczamy średnia dla kazdego produktu:
suma_cen = {
    produkt: sum(wartosc)
     for produkt, wartosc in grouped.items()
}

print("Suma cen produktów")
print(suma_cen)
```

```python
"""Zadanie 11 – Feature engineering z comprehension
Masz dane klientów: wiek i dochód. Dodaj nowe cechy: kat_wiekowa (młody/
średni/starszy) i kat_dochodu (niski/średni/wysoki).
Dataset:
klienci = [
{"id": 1, "wiek": 25, "dochod": 4500},
{"id": 2, "wiek": 45, "dochod": 9200},
{"id": 3, "wiek": 32, "dochod": 6800},
{"id": 4, "wiek": 28, "dochod": 5200},
{"id": 5, "wiek": 55, "dochod": 12000}
]
Wymagania:
Kategoria wiekowa: <30 = młody, 30-50 = średni, >50 = starszy
Kategoria dochodu: <5000 = niski, 5000-10000 = średni, >10000 = wysoki
Użyj list comprehension do tworzenia nowej listy z dodatkowymi cechami
Oczekiwany rezultat:
Lista klientów z 2 nowymi cechami
"""

klienci_raw = [
{"id": 1, "wiek": 25, "dochod": 4500},
{"id": 2, "wiek": 45, "dochod": 9200},
{"id": 3, "wiek": 32, "dochod": 6800},
{"id": 4, "wiek": 28, "dochod": 5200},
{"id": 5, "wiek": 55, "dochod": 12000}
]

def kategoria_wiekowa(wiek):
    if wiek < 30:
        return "mlody"
    elif wiek < 50:
        return "sredni"
    else:
        return "starszy"
    
def poziom_dochodu(dochod):
    if dochod < 5000:
        return "niski"
    elif dochod < 8000:
        return "wysoki"
    else:
        return "wysoki"
    
#Tworzymy nowe cechy uzywajac comprehension    
klienci_engineered = [
    {
        "id": k["id"],
        "wiek": k["wiek"],
        "dochod": k["dochod"],
        "kat_wiekowa": kategoria_wiekowa(k["wiek"]),
        "poz_dochodu": poziom_dochodu(k["dochod"]),
    }
    for k in klienci_raw
]

print("Klienci z dodatkowymi cechami:")
for klient in klienci_engineered:
    print(klient)
```
