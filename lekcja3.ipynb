```python
"""Zadanie 1 – Kalkulator BMI
Napisz program, który obliczy wskaźnik BMI (Body Mass Index).
Dane:
Wymagania:
Oczekiwany rezultat:
(proste)
Waga: 70 kg
Wzrost: 1.75 m
Stwórz zmienne waga i wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost²)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)
BMI: 22.86
Interpretacja: Norma"""

waga = 70
wzrost = 1.75

BMI = waga/ (wzrost**2)
print(f'BMI {BMI:.2f}')
if BMI < 18.5:
    print('niedowaga')
elif BMI < 25:
    print("norma")
else:
    print('nadwaga')
```

```python
"""Zadanie 2 – Konwersja temperatur
Napisz program konwertujący temperatury Celsius na Fahrenheit i Kelvin.
Dane:
Wymagania:
Oczekiwany rezultat:
(proste)
Temperatura: 23.5°C
Fahrenheit = Celsius × 9/5 + 32
Kelvin = Celsius + 273.15
Wyświetl wszystkie 3 wartości sformatowane z 1 miejscem po przecinku
23.5°C = 74.3°F = 296.6 K"""

# Dane wejściowe
celsius = 23.5

# Obliczenia
fahrenheit = celsius * 9/5 + 32
kelvin = celsius + 273.15

# Wyświetlenie wyniku z formatowaniem do 1 miejsca po przecinku
print(f"{celsius:.1f}°C = {fahrenheit:.1f}°F = {kelvin:.1f} K")
```

```python
"""Zadanie 3 – Podział zbioru danych
Oblicz rozmiary zbiorów treningowego, walidacyjnego i testowego.
Dane:
Wymagania:
(proste)
Całkowita liczba próbek: 1000
Proporcje: 70% trening, 15% walidacja, 15% test
Użyj mnożenia i dzielenia całkowitego
Upewnij się, że suma = 1000 (obsłuż resztę z dzielenia)
Wyświetl rozmiary wszystkich zbiorów"""

data = 1001
training_data = data * 0.7
validaton_data = data * 0.15
test_data = data * 0.15

print(f"Data: {data}, validation: {validaton_data: .2f}, training: {training_data:.2f}, validation: {validaton_data: .2f}, test: {test_data:.0f}")
print(f"Original data is equal to sum of calculations: {data == (training_data+validaton_data+test_data)}")

def calculate(data):
    training_data = data * 0.7
    validaton_data = data * 0.15
    test_data = data * 0.15
    my_object = {
        "training_data": training_data,
        "validaton_data": validaton_data,
        "test_data": test_data
    }
    return my_object

result = calculate(data)
print(result)
```

```python
"""Zadanie 9 – Normalizacja Min-Max
Znormalizuj wartość do zakresu [0, 1].
Dane:
Wartość: 75
Wymagania:
(średnie)
Minimum w kolumnie: 50
Maximum w kolumnie: 100
Wzór: normalized = (value - min) / (max - min)
Oblicz dla wartości 75, 50, 100, 62.5
Sprawdź, czy wynik mieści się w [0, 1]
Wyświetl tabelkę: wartość oryginalna | wartość znormalizowana"""

value = 75
min_in_kolumn = 50
max_in_kolumn = 100

value_origin = [75, 50, 100, 62.5]
print("Input data:", value_origin)

value_normalized = []

for val in value_origin:
    normalized = (val - min_in_kolumn) / (max_in_kolumn - min_in_kolumn)
    if 0.0 <= normalized <= 1.0:
        print(f"Status: OK -> {normalized} is within [0, 1] range")
    else:
        print(f"Status: WARNING -> {normalized} is out of bounds!")
    value_normalized.append(normalized) # adding the result to a new container


print("value origin | value normalized") 
for origin, normalized in zip(value_origin, value_normalized):
    print(f"{origin} | {normalized}")
```
