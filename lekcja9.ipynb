```python
"""Zadanie 3 – Klasa BankAccount
Stwórz klasę BankAccount z atrybutami: wlasciciel , saldo .
Metody:
wplac(kwota) – zwiększa saldo
wyplac(kwota) – zmniejsza saldo (tylko jeśli wystarczy środków)
pokaz_saldo() – wyświetla saldo

Stwórz kilka kont i przetestuj operacje.
"""

class BankAccount:
    def __init__(self, wlasciciel: str, saldo: int):
        self.wlasciciel = wlasciciel
        self.saldo = saldo
    
    def wplac(self, kwota):
        if kwota > 0:
            self.saldo += kwota
            return True
        return False

    def wyplac(self, kwota) -> bool:
        if 0 < kwota <= self.saldo:
            self.saldo -= kwota
            return True
        return False
    
    def pokaz_saldo(self):
        # To jedyne miejsce, gdzie zazwyczaj zostawiamy printa do testów
        print(f"Konto: {self.wlasciciel} | Saldo: {self.saldo} zł")

# Testowanie - Logika wyświetlania jest poza klasa!
konto1 = BankAccount("Jan Kowalski", 1000)

# Teraz "system" podejmuje decyzję, co zrobić z wynikiem:
czy_dalo_sie = konto1.wyplac(1500)

if czy_dalo_sie:
    print("Bankomat wypłaca banknoty.")
else:
    print("Czerwony komunikat na ekranie banknotu: Brak środków!")
```

```python
"""Zadanie 5 – Klasa Student
Stwórz klasę Student z atrybutami: imie , numer_indeksu , oceny (lista ocen).
Metody:
- dodaj_ocene(ocena) – dodaje ocenę do listy
- srednia() – zwraca średnią ocen
- __str__() – reprezentacja tekstowa
Stwórz 3 studentów, dodaj im oceny i wyświetl średnie.
"""

class Student:
    def __init__ (self, imie: str, numer_indeksu: int):
        self.imie = imie
        self.numer_indeksu = numer_indeksu
        self.oceny = []
    
    def dodaj_ocene(self, ocena: int):
        self.oceny.append(ocena)

    def srednia(self):
        if len(self.oceny) == 0:
            return 0
        return sum(self.oceny) / len(self.oceny)

    def __str__(self):
        return f"imie: {self.imie}, numer indeksu: {self.numer_indeksu}, średnia: {self.srednia():.2f}"
    
# Testowanie:
# Pierwszy student
student1 = Student(imie = 'Anna Maria', numer_indeksu = 1223)
student1.dodaj_ocene(4)
student1.dodaj_ocene(2)
student1.dodaj_ocene(3)
print(student1)
# Drugi student
student2 = Student(imie = 'Jan Kowalski', numer_indeksu = 1230)
student2.dodaj_ocene(5)
student2.dodaj_ocene(5)
student2.dodaj_ocene(4)
print(student2)
# Trzeci student
student3 = Student(imie='Piotr Nowak', numer_indeksu=1240)
student3.dodaj_ocene(3)
student3.dodaj_ocene(3)
student3.dodaj_ocene(2)
print(student3)
```

```python
"""Zadanie 7 – Klasa Pracownik
Stwórz klasę Pracownik z atrybutami: imie , stanowisko , pensja .
Metody:
podwyzka(procent) – zwiększa pensję o procent%
__str__() – reprezentacja tekstowa
Stwórz 3 pracowników i daj im podwyżki.
"""

class Pracownik:
    def __init__(self, imie: str, stanowisko: str, pensja: int):
        self.imie = imie
        self.stanowisko = stanowisko
        self.pensja = pensja
    
    def podwyzka(self, procent: int):
        if procent > 0:
            # Obliczamy kwotę podwyżki i trwale dodajemy do obecnej pensji
            kwota_podwyzki = self.pensja * (procent / 100)
            self.pensja += kwota_podwyzki
            self.pensja = round(self.pensja, 2)
    
    def __str__(self):
        return f"Imię: {self.imie}, Stanowisko: {self.stanowisko}, Pensja: {self.pensja} PLN"

#Testy:
p1 = Pracownik(imie = 'Anna Maria', stanowisko = 'nauczyciel', pensja = 5000)
p2 = Pracownik(imie = 'Jan Kowalski', stanowisko ='dyrektor', pensja = 10000)
p3 = Pracownik(imie = 'Michał Michalski', stanowisko = 'woźny', pensja = 4500)

print("Przed podwyzka")
print(p1)
print(p2)
print(p3)

p1.podwyzka(10)
p2.podwyzka(15)
p3.podwyzka(5)

print("\n Po podwyzce")
print(p1)
print(p2)
print(p3)
```

```python
"""Zadanie 9 – Pipeline z wieloma krokami
Stwórz 4 komponenty:
- FiltrowanieDodatnich – usuwa liczby ujemne
- Zaokraglanie – zaokrągla do całkowitych
- Skalowanie – mnoży przez stały współczynnik
- Sumowanie – zwraca sumę listy
Stwórz klasę Pipeline i przetestuj na danych: [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]
"""
# 4 komponenty:
class FiltrowanieDodatnich():
    def wykonaj(self, dane):
        return list(filter(lambda l: l > 0, dane))
    
class Zaokraglanie():
    def wykonaj(self, dane):
        return list(map(lambda l: round (l), dane))
    
class Skalowanie():
    def __init__ (self, wspolczynnik):
        self.wspolczynnik = wspolczynnik

    def wykonaj(self, dane):
        return list(map(lambda l: l * self.wspolczynnik, dane))

class Sumowanie:
    def wykonaj(self, dane):
        return sum(dane)

# Pipeline:
class Pipeline:
    def __init__(self):
        self.kroki = []
    
    def dodaj_krok(self, krok):
        self.kroki.append(krok)
    
    def uruchom(self, dane):
        wynik = dane
        print(f"Dane na starcie: {wynik}")

        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
            print(f"Po przejściu przez{krok.__class__.__name__}: {wynik}")
        return wynik
    
# Testowanie
dane_testowe = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

test_pipeline = Pipeline()

test_pipeline.dodaj_krok(FiltrowanieDodatnich())
test_pipeline.dodaj_krok(Zaokraglanie())
test_pipeline.dodaj_krok(Skalowanie(10))
test_pipeline.dodaj_krok(Sumowanie())

print("Start Pipeline'a")
wynik_ostateczny = test_pipeline.uruchom(dane_testowe)

print(f"Ostateczny wynik z Pipeline'a to: {wynik_ostateczny}")
```

```python
"""Zadanie 10 – Klasa Zamowienie z kompozycją
Stwórz klasę Produkt (nazwa, cena) oraz klasę Zamowienie , która zawiera listę
produktów.
Metody Zamowienie :
dodaj_produkt(produkt, ilosc)
laczna_wartosc() – suma wartości zamówienia
pokaz_zamowienie() – wyświetla wszystkie produkty i łączną wartość

Stwórz zamówienie z 5 produktami.
"""

class Produkt():
    def __init__ (self, nazwa, cena):
        self.nazwa = nazwa
        self.cena = cena
    
class Zamowienie():
    def __init__ (self):
        self.lista_produktow = []

    def dodaj_produkt(self, produkt, ilosc):
        wartosc = {
            "produkt": produkt,
            "ilosc": ilosc
        }
        if ilosc > 0:
            self.lista_produktow.append(wartosc)

    def laczna_wartosc(self):
        suma = 0
        for element in self.lista_produktow:
            p = element["produkt"]
            sztuki = element["ilosc"]
            suma += p.cena * sztuki
        return suma

    def pokaz_zamowienie(self):
        print("Zamówienie")

        for element in self.lista_produktow:
            p = element["produkt"]
            sztuki = element["ilosc"]

            print(f"Nazwa produktu: {p.nazwa} | Ilość: {sztuki} | Wartość: {p.cena * sztuki} zł")

        suma_calkowita = self.laczna_wartosc()
        print(f"ŁACZNA WARTOŚĆ: {suma_calkowita} zł")

#Testowanie zadania:

# 1. Tworzymy produkty w sklepie:
chleb = Produkt("Chleb Wiejski", 6)
maslo = Produkt("Masło Łaciate", 8)
mleko = Produkt("Mleko Piatnica", 4)

# 2. Bierzemy pusty koszyk (tworzymy zamówienie)
zakupy = Zamowienie()

#3. Wrzucamy produkty do koszyka
zakupy.dodaj_produkt(chleb, 2)
zakupy.dodaj_produkt(maslo, 1)
zakupy.dodaj_produkt(mleko, 3)

#4. Idziemy do kasy (wyświetlamy paragon)
zakupy.pokaz_zamowienie()
```

```python
"""Zadanie 12 – Dziedziczenie – FiguraGeometryczna
Stwórz klasę bazową Figura z metodą pole() .
Stwórz klasy potomne:
- Prostokat(szerokosc, wysokosc)
- Kolo(promien
Każda klasa implementuje pole() .
Stwórz listę figur i oblicz łączne pole.
"""

import math
from abc import abstractmethod, ABC

class Figura(ABC):
    def pole(self):
        pass

class Prostokat(Figura):
    def __init__ (self, szerokosc, wysokosc):
        self.szerokosc = szerokosc
        self.wysokosc = wysokosc
    
    def pole(self):
        return self.szerokosc * self.wysokosc

class Kolo(Figura):
    def __init__ (self, promien):
        self.promien = promien

    def pole(self):
        return math.pi * self.promien ** 2

# Lista figur testowych:
testowy_prostokat = Prostokat(szerokosc=4, wysokosc=5)
testowe_kolo = Kolo(promien=3)
kolejny_prostokat = Prostokat(szerokosc=2, wysokosc=10)

lista_figur = [testowy_prostokat, testowe_kolo, kolejny_prostokat]

calkowite_pole = 0

for element in lista_figur:
    calkowite_pole += element.pole()

print(f"Łaczne pole wszystkich figur wynosi: {calkowite_pole:.2f}")
```
