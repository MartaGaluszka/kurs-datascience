```python
"""Zadanie 1 – Tworzenie i indeksowanie listy
Stwórz listę zawierającą pomiary temperatury z 7 dni: [22.5, 23.0, 21.5, 24.0, 23.5,
22.0, 23.0] . 
Wyświetl:
Temperaturę z pierwszego dnia
Temperaturę z ostatniego dnia
Temperatury z trzech środkowych dni (slicing)
"""

temp = [22.5, 23.0, 21.5, 24.0, 23.5,
22.0, 23.0]

print(f"Temperature on the first day: {temp[0]}")
print(f"Temperature on the last day: {temp[-1]}")
print(f"Temperature of three average temperatures: {temp[2:5]}")
```

```python
"""Zadanie 2 – Dodawanie elementów do listy
Zacznij od pustej listy results = [] . Dodaj do niej 5 wartości accuracy: 0.75, 0.82, 0.79,
0.85, 0.88. Następnie:
Wyświetl długość listy
Wyświetl średnią wartość
Wyświetl maksymalną wartość"""

results = []
accuracy = [0.75, 0.82, 0.79, 0.85, 0.88]
results.extend(accuracy)

print(f"List: {results}")
print(f"List length: {len(results)}")
print(f"Average value: {sum(results) / len(results)}")
print(f"Maximum value: {max(results)}")
```
