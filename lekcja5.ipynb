```python
"""Zadanie 1 – Tworzenie i indeksowanie listy
Stwórz listę zawierającą pomiary temperatury z 7 dni: [22.5, 23.0, 21.5, 24.0, 23.5,
22.0, 23.0] . 
Wyświetl:
Temperaturę z pierwszego dnia
Temperaturę z ostatniego dnia
Temperatury z trzech środkowych dni (slicing)

Create a list containing temperature measurements from 7 days: [22.5, 23.0, 21.5, 24.0, 
23.5, 22.0, 23.0]. Display: The temperature on the first day, 
The temperature on the last day, The temperatures of the three middle days (slicing).
"""
# List containing 7 days of temperature measurements
temp = [22.5, 23.0, 21.5, 24.0, 23.5, 22.0, 23.0]

# Index 0 accesses the first element
print(f"Temperature on the first day: {temp[0]}")

# Index -1 accesses the last element (negative indexing)
print(f"Temperature on the last day: {temp[-1]}")

# Slicing [2:5] extracts elements at index 2, 3, and 4 (the three middle days)
# Fixed phrasing: 'middle days' instead of 'average temperatures'
print(f"Temperature of three average temperatures: {temp[2:5]}")
```

```python
"""Zadanie 2 – Dodawanie elementów do listy
Zacznij od pustej listy results = [] . Dodaj do niej 5 wartości accuracy: 0.75, 0.82, 0.79,
0.85, 0.88. Następnie:
Wyświetl długość listy
Wyświetl średnią wartość
Wyświetl maksymalną wartość

Start with an empty list results = []. Add 5 accuracy values to it: 0.75, 0.82, 0.79, 
0.85, 0.88. Then: Display the length of the list, 
Display the average value, Display the maximum value.
"""

# Initialize an empty list
results = []

# List of new accuracy values to add
accuracy = [0.75, 0.82, 0.79, 0.85, 0.88]

# Use extend() to add multiple elements to the end of the list
results.extend(accuracy)

print(f"List: {results}")

# len() returns the number of elements in the list
print(f"List length: {len(results)}")

# Calculate the average (mean) by dividing the sum by the length
print(f"Average value: {sum(results) / len(results)}")

# max() returns the highest value in the list
print(f"Maximum value: {max(results)}")
```

```python
"""Zadanie 3 – Sortowanie listy
Masz listę wyników: [0.78, 0.92, 0.85, 0.88, 0.81] . Posortuj ją malejąco i wyświetl 3
najlepsze wyniki.

Task 3 – Sorting a list
You have a list of results: [0.78, 0.92, 0.85, 0.88, 0.81]. 
Sort it in descending order and display the top 3 results.
"""
# List of raw results (e.g., model evaluation scores)
results = [0.78, 0.92, 0.85, 0.88, 0.81]

# Sort the list in descending order (from highest to lowest) in place
results.sort(reverse=True) 

# Print the fully sorted list
print(f"List sorted by highest value: {results}")

# Print the top 3 results using list slicing (fixed grammar)
print(f"Three the biggest results: {results[:3]} ")
```

```python
"""Zadanie 4 – Czyszczenie nazw kolumn
Masz listę nazw kolumn: [" Customer ID ", "TOTAL AMOUNT", " date "] . Wyczyść je
zgodnie z konwencją:
Usuń białe znaki (strip)
Zamień na małe litery (lower)
Zamień spacje na podkreślniki (replace)

Task 4 – Cleaning column names
You have a list of column names: [" Customer ID ", "TOTAL AMOUNT", " date "]. 
Clean them according to the convention: Remove whitespaces (strip), 
Convert to lowercase (lower), Replace spaces with underscores (replace).
"""

# List of raw column names, typically extracted from a messy dataset
cols = [" Customer ID ", "TOTAL AMOUNT", " date "]

# Use list comprehension and method chaining to clean the column names:
# 1. strip() removes leading and trailing whitespaces
# 2. lower() converts all characters to lowercase
# 3. replace(" ", "_") replaces any remaining internal spaces with underscores
clean_cols = [c.strip().lower().replace(" ", "_") for c in cols]

# Print the final list of cleaned column names (fixed typo: 'columns')
print(f"Clean colums: {clean_cols}")
```

```python
"""Zadanie 5 – Parsowanie CSV
Masz string CSV: "Alice,28,Data Scientist,75000" . Użyj split(',') aby wyciągnąć:
Imię
Wiek (jako int)
Zawód
Pensję (jako int)
Wyświetl te informacje w formacie: "Alice ma 28 lat, pracuje jako Data Scientist i
zarabia $75000"

Task 5 – Parsing CSV
You have a CSV string: "Alice,28,Data Scientist,75000". Use split(',') to extract: 
Name, Age (as int), Occupation, Salary (as int). Display this information in the format: 
"Alice is 28 years old, works as a Data Scientist and earns $75000".
"""

# Define the CSV (Comma-Separated Values) string
csv_data = "Alice,28,Data Scientist,75000"

# Split the string into a list of substrings using the comma as a delimiter
parts = csv_data.split(',')

# Extract each part by its index
name = parts[0]

# Cast age and salary to integers (int) for numerical representation
age = int(parts[1])
occupation =parts[2]
salary = int(parts[3])

# Print the formatted string (fixed typos: 'years', 'works', and added '$')
print(f"{name} is {age} years old, works as {occupation} and earns ${salary}.")
```

```python
"""Zadanie 6 – Formatowanie raportu
Masz dane: model_name = "XGBoost" , accuracy = 0.8934 , time = 45.67 . Użyj f-
string aby stworzyć raport:
(proste)
Model: XGBoost
Accuracy: 89.34%
Training time: 45.67s

Task 6 – Formatting a report
You have the following data: model_name = "XGBoost", accuracy = 0.8934, 
time = 45.67. Use an f-string to create a report.
"""

# Define the machine learning model parameters
model_name = "XGBoost"
accuracy = 0.8934
time = 45.67

# Print the report using f-strings for dynamic formatting
print(f"Model: {model_name}")

# Use the percentage format specifier (:.2%) to format the float as a percentage with 2 decimal places
print(f"Accuracy: {accuracy:.2%}") 

# Append 's' directly to the f-string for seconds
print(f"Training time: {time}s")
```

```python
"""Zadanie 7 – Zliczanie wystąpień
Masz listę predykcji: ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam'] .
Policz ile jest 'spam' i ile 'ham' używając metody count()

Task 7 – Counting occurrences
You have a list of predictions: ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam'].
Count how many times 'spam' and 'ham' appear using the count() method.
"""
# List of predictions (e.g., from a machine learning classification model)
preds = ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam']

# Count the occurrences of 'spam' in the list and print the result
print(f"Spam: {preds.count('spam')}")

# Count the occurrences of 'ham' in the list and print the result
print(f"Ham: {preds.count('ham')}")
```

```python
"""Zadanie 8 – Ekstakcja informacji ze stringa
Masz nazwę pliku: "model_results_2024_Q3.csv" . Wyciągnij rok i kwartał używając
slicing lub split.

You have a filename: "model_results_2024_Q3.csv". Extract the year and quarter using slicing or split.
"""

filename = "model_results_2024_Q3.csv"

# Split the filename by underscores to isolate the components
parts = filename.split('_') 

# Extract the year (it's the 3rd element in the resulting list, so index 2)
year = parts[2]

# Extract the quarter by taking the 4th element, splitting it by a dot, and grabbing the first part
quarter = parts[3].split('.')[0]

# Print the final results using an f-string
print(f"Year: {year}, Quater: {quarter}")
```

```python
"""Zadanie 9 – Filtrowanie outlierów
Masz listę pomiarów: [23, 24, 22, 150, 23, 25, -10, 24, 23] . Outlierem jest wartość
poza zakresem [20, 30]. Stwórz nową listę bez outlierów używając list comprehension.

Task 9: Filtering outliers
GOAL: Remove outliers (values outside the 20-30 range) from the dataset using list comprehension.
"""


# 1. Raw data including outliers (like 150 and -10)
measurements = [23, 24, 22, 150, 23, 25, -10, 24, 23]

# 2. Filter data using list comprehension with an 'if' condition
# We iterate through 'measurements' and keep 'val' ONLY IF it is between 20 and 30
valid_measurements = [val for val in measurements if 20 <= val <= 30]

# 3. Display the results
print(f"Original data: {measurements}")
print(f"Cleaned data: {valid_measurements}")
```

```python
"""Zadanie 10 – Agregacja danych po kategoriach
Masz listę transakcji (kategoria, kwota):
transactions = [('food', 50),('transport', 20),('food', 30),('entertainment', 40),('transport', 15),('food', 25)]
Oblicz łączną kwotę dla każdej kategorii. Wynik zapisz w słowniku (użyj pętli i warunków).

Task 10: Data aggregation by category
GOAL: Calculate the total amount for each category. Save the result in a dictionary.
"""

# 1. List of transaction (tuples containing: category, amount) 
transactions = [
('food', 50),
('transport', 20),
('food', 30),
('entertainment', 40),
('transport', 15),
('food', 25),
]

# 2. Create an emoty dictionary to store the totals
category_totals = {}

# 3. Loop through each transaction 
for category, amount in transactions:
    # If the category already exists in our dictionary
    if category in category_totals:
        category_totals[category] += amount # Add the current to the existing total

    # If the category is NOT in our dictionary yet (first time we see it)
    else:
        category_totals[category] = amount # Create a new entry with the first 

#4. Print the final report
print("Expense Summary:")
print("-" *25)

# .items() extracts both the category name and its total value
for category, total in category_totals.items():
    print(f"{category.title():<15} | ${total}")
```
