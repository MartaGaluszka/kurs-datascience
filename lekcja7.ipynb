```python
"""Zadanie 1 – Klasyfikacja liczb parzystych i nieparzystych
Napisz program, który dla listy liczb [12, 7, 23, 45, 8, 19, 34, 56, 91, 2] wyświetli,
które liczby są parzyste, a które nieparzyste. Użyj instrukcji if/else i operatora modulo
% .
Wymagania:
Iteruj po liście używając pętli for
Dla każdej liczby sprawdź: if liczba % 2 == 0
Wyświetl wynik w formacie: "12 - parzysta", "7 - nieparzysta"
Oczekiwany rezultat:
Lista 10 linii z klasyfikacją każdej liczby"""

numbers = [12, 7, 23, 45, 8, 19, 34, 56, 91, 2]

list2 = [f"{number} - parzysta" if number % 2 == 0 else f"{number} - nieparzysta" for number in numbers]
print(list2)
```

```python
"""Zadanie 2 – Suma liczb dodatnich
Masz listę temperatury: [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3] . Oblicz sumę tylko
dodatnich wartości używając pętli for i instrukcji warunkowej.
Wymagania:
Użyj zmiennej suma = 0
W pętli sprawdzaj if temp > 0:
Dodaj tylko dodatnie wartości do sumy
Oczekiwany rezultat:
Wyświetlenie sumy dodatnich temperatur
Liczba dodatnich pomiarów"""

temperatury = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma = 0
pomiar = 0

for temperatura in temperatury:
    if temperatura > 0:
        suma += temperatura
        pomiar += 1

print(suma)
print(pomiar)
```

```python
"""Zadanie 3 – Odliczanie z krokiem
Napisz program, który wyświetli odliczanie od 20 do 0 z krokiem -2 
(20, 18, 16, ..., 2, 0). Użyj pętli for i funkcji range() .
Wymagania:
for i in range(20, -1, -2):
Wyświetl każdą liczbę
Oczekiwany rezultat:
Lista liczb: 20, 18, 16, 14, 12, 10, 8, 6, 4, 2, 0"""

for i in range(20, -1, -2):
    print(i)
```

```python
"""Zadanie 4 – Znajdź maksimum ręcznie
Masz listę wyników testów: [78, 92, 65, 88, 95, 71, 85] . Znajdź 
maksymalny wynik używając pętli for (bez funkcji max() ).
Wymagania:
Inicjuj maksimum = wyniki[0]
W pętli porównuj: if wynik > maksimum:
Aktualizuj maksimum gdy znajdziesz większą wartość
Oczekiwany rezultat:
Maksymalny wynik: 95"""

wyniki = [78, 92, 65, 88, 95, 71, 85]

def obliczenia(input_arr):

    maximum = input_arr[0]

    for value in input_arr:
        if value > maximum:
            maximum = value 
    return maximum

print(f'Maximum wynosi: {obliczenia(wyniki)}')   

```

```python
"""Zadanie 5 – Suma liczb w zakresie
Oblicz sumę wszystkich liczb od 1 do 50 używając pętli while .
Wymagania:
Zmienne: suma = 0 , i = 1
while i <= 50:
Dodaj i do sumy i zwiększ i += 1
Sprawdź wynik wzorem: n*(n+1)/2 = 50*51/2 = 1275
Oczekiwany rezultat:
Suma: 1275"""

suma = 0
i = 1

while i <= 50:
    suma += i
    i += 1

print(f"Suma liczb od 1 do 50: {suma}")
# Wzór sprawdzajcy: n*(n+1)/2 = 50*51/2 = 1275
```

```python
"""Zadanie 6 – Filtrowanie listy z warunkiem
Masz listę temperatur ciała: [36.5, 37.8, 36.2, 38.5, 36.9, 39.2, 37.0] . Wyświetl
tylko te pomiary, które wskazują gorączkę (temperatura >= 38.0°C).
Wymagania:
Pętla for po liście
Instrukcja if temperatura >= 38.0:
Wyświetl temperaturę z ostrzeżeniem
Oczekiwany rezultat:
Lista temperatur >= 38.0 z komunikatem 'Gorączka'"""

temperatury = [36.5, 37.8, 36.2, 38.5, 36.9, 39.2, 37.0]
wysokie_temp = []
komunikat = 'Goraczka'

for temperatura in temperatury:
    if temperatura >= 38.0:
        wysokie_temp.append(temperatura)

print(f"Lista temperatur z gorączką: {wysokie_temp} - Gorączka")
```

```python
"""Zadanie 7 – Tabliczka mnożenia
Wyświetl tabliczkę mnożenia dla liczby 7 (od 7×1 do 7×10) 
używając pętli for i range() .
Wymagania:

for i in range(1, 11):
Oblicz wynik = 7 * i
Wyświetl: "7 × 1 = 7"
Oczekiwany rezultat:
10 linii z wynikami mnożenia"""

for i in range(1,11):
    wynik = 7 * i
    print(f"7 x {i} = {wynik}")
```

```python
"""Zadanie 8 – Konwersja listy używając list comprehension
Masz listę wzrostów w centymetrach: [165, 178, 182, 159, 171, 190] . 
Przekonwertuj je na metry używając list comprehension.
Wymagania:
wzrosty_metry = [wzrost / 100 for wzrost in wzrosty_cm]
Wyświetl obie listy
Oczekiwany rezultat:
Lista wzrostów w metrach: [1.65, 1.78, 1.82, 1.59, 1.71, 1.90]"""

wzrosty_cm = [165, 178, 182, 159, 171, 190]
wzrosty_metry = [wzrost / 100 for wzrost in wzrosty_cm]

print(f"Lista wzrostów w centymetrach: {wzrosty_cm}")
print(f"Lista wzrostów w metrach: {wzrosty_metry}")
```

```python
"""Zadanie 9 – Kategorie BMI dla listy pacjentów
Masz dane pacjentów: wagi = [70, 85, 62, 95, 78] (kg) i wzrosty = [1.75, 1.80,
1.65, 1.78, 1.72] (m). Oblicz BMI dla każdego i sklasyfikuj (niedowaga < 18.5, norma
18.5-24.9, nadwaga >= 25).
Dataset:
Wymagania:
Listy równoległe: wagi i wzrosty
Użyj pętli for i in range(len(wagi)):
Oblicz bmi = wagi[i] / (wzrosty[i] ** 2)
Sklasyfikuj używając if/elif/else
Wyświetl wyniki dla każdego pacjenta
Oczekiwany rezultat:
5 linii z BMI i kategorią dla każdego pacjenta"""

wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72]

for i in range(len(wagi)):
    bmi = wagi[i] / (wzrosty[i] ** 2)
    if bmi < 18.5:
        print(f"{bmi:.2f} - niedowaga")
    
    elif bmi < 25:
        print(f"{bmi:.2f} - norma")
    
    else:
        print(f"{bmi:.2f} - nadwaga")
```

```python
"""Zadanie 10 – Znajdź wszystkie liczby pierwsze do 100
Napisz program, który znajdzie wszystkie liczby pierwsze od 2 do 100 używając
zagnieżdżonych pętli.
Wymagania:
Pętla zewnętrzna: for n in range(2, 101):
Pętla wewnętrzna: sprawdź dzielniki od 2 do √n
Zmienna czy_pierwsza = True
Jeśli znajdziesz dzielnik, ustaw czy_pierwsza = False i użyj break
Jeśli czy_pierwsza == True , dodaj do listy
Oczekiwany rezultat:
Lista liczb pierwszych: [2, 3, 5, 7, 11, 13, ..., 97]
Liczba znalezionych: 25"""

liczby_pierwsze = []

for n in range(2, 101):
    czy_pierwsza = True
    
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            czy_pierwsza = False
            break

    if czy_pierwsza == True:
        liczby_pierwsze.append(n)

print("Lista licz pierwszych" ,liczby_pierwsze)
print("Liczba znalezionych:", len(liczby_pierwsze))
```

```python
"""Zadanie 11 – Analiza trendów wzrostowych/spadkowych
Masz dane sprzedaży miesięcznej: [1200, 1350, 1290, 1480, 1520, 1510, 1680,
1720, 1700, 1890, 1950, 2100] . Zidentyfikuj miesiące, w których sprzedaż wzrosła w
porównaniu do poprzedniego miesiąca.
Dataset:
Wymagania:
Lista 12 wartości (sprzedaż miesięczna w PLN)
Pętla for i in range(1, len(sprzedaz)):
Porównaj sprzedaz[i] z sprzedaz[i-1]
Oblicz zmianę procentową
Wyświetl miesiące ze wzrostem
Oczekiwany rezultat:
Lista miesięcy ze wzrostem i procentowy przyrost"""

# Dataset:
sprzedaz = [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100]
miesiace_ze_wzrostem = []

for i in range(1, len(sprzedaz)):
    if sprzedaz[i] > sprzedaz[i-1]:
        procentowy_przyrost = (((sprzedaz[i]-sprzedaz[i-1]) / sprzedaz[i-1]) * 100)
        wiadomosc_przyrost_procentowy = (f"{procentowy_przyrost:.2f} %")
        miesiace_ze_wzrostem.append((i+1, wiadomosc_przyrost_procentowy))

print(f"Miesiące ze wzrostem (Miesiąc, Przyrost): {miesiace_ze_wzrostem}")
```
