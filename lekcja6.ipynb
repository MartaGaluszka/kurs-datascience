```python
"""
Zadanie 9 – Inwersja słownika
Masz słownik mapujący nazwy produktów na kategorie: {'laptop': 'elektronika',
'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'} . Stwórz
odwrotny słownik gdzie klucze to kategorie, a wartości to listy produktów.
Wymagania:
Iteruj po oryginalnym słowniku
Zgrupuj produkty według kategorii
Oczekiwany wynik:
Wynik: {'elektronika': ['laptop', 'klawiatura'], 'meble': [...]}
Słownik kategoria: lista_produktów

You have a dictionary mapping product names to categories: 
{'laptop': 'elektronika', 'klawiatura': 'elektronika', 'krzesło': 'meble',
 'biurko': 'meble'}. Create an inverted dictionary where the keys are categories, 
 and the values are lists of products.
Requirements: Iterate through the original dictionary. Group products by category.
"""

# 1. Original dictionary mapping product to categories:
products = {
    'laptop': 'elektronika',
    'klawiatura': 'elektronika', 
    'krzesło': 'meble', 
    'biurko': 'meble'
}

# 2. Initialize an empty dictionary for our grouped results
inverted_dict = {}

# 3. Iterate through both keys (product) and values (category) using .items()
for product, category in products.items():
    #If the category in not yet a key in our new dictonary, add it with an empty list
    if category not in inverted_dict:
        inverted_dict[category] = []

    #Append the product to the correct category list
    inverted_dict[category].append(product)

# 4. Print the final inverted dictionary
print(f"Inverted dictionary: {inverted_dict}") 
```

```python
"""Zadanie 10 – Walidacja danych ze zbiorem dozwolonych wartości
Masz listę kategorii od użytkowników: ['sport', 'muzyka', 'SPORT', 'film', 'xyz',
'nauka', 'gaming'] . Dozwolone kategorie to {'sport', 'muzyka', 'film', 'nauka',
'gry'} . Znajdź nieprawidłowe kategorie (uwzględnij wielkość liter - konwertuj na małe).
Wymagania:
Konwertuj wszystkie kategorie na małe litery
Użyj operacji zbiorowych by znaleźć nieprawidłowe
Wypisz prawidłowe i nieprawidłowe osobno

Task 10 – Data validation with a set of allowed values
You have a list of categories from users: ['sport', 'muzyka', 'SPORT', 'film', 'xyz', 
'nauka', 'gaming']. The allowed categories are {'sport', 'muzyka', 'film', 'nauka', 'gry'}. 
Find the invalid categories (handle case sensitivity - convert to lowercase).
Requirements: Convert all categories to lowercase. Use set operations to find invalid ones. 
Print valid and invalid categories separately.
"""

# 1. Input data 
user_categories = ['sport', 'muzyka', 'SPORT', 'film', 'xyz', 'nauka', 'gaming']
allowed_categories = {'sport', 'muzyka', 'film', 'nauka', 'gry'}

# 2. Convert all user categories to lowercase and turn them into a SET
normalid_categories = {category.lower() for category in user_categories}
print(f"normalid categories: {normalid_categories}")

# 3. Use set operation to find invalid and valid categories
invalid_categories = normalid_categories - allowed_categories
print(f"invalid categories: {invalid_categories}")

#VALID: The intersection (common part) of user categories and allowed categories
valid_categories = normalid_categories & allowed_categories

# 4. Print the results
print("Print the results")
print(f"Valid categories: {valid_categories}")
print(f"Invalid categories: {invalid_categories}")
```

```python
"""Zadanie 11 – Dictionary comprehension - kwadraty liczb
Stwórz słownik używając dictionary comprehension, gdzie klucze to liczby od 1 do 20, a
wartości to ich kwadraty. Następnie przefiltruj ten słownik (też comprehension!) by zawierał
tylko liczby parzyste.
Wymagania:
Użyj dictionary comprehension 2 razy
Pierwszy słownik: wszystkie liczby 1-20
Drugi słownik: tylko parzyste klucze
Oczekiwany wynik:
Dwa słowniki wypisane

Task 11 – Dictionary comprehension - squares of numbers
Create a dictionary using dictionary comprehension, where keys are numbers from 1 to 20, 
and values are their squares. Then filter this dictionary (also using comprehension!) so
 it contains only even numbers.
Requirements: Use dictionary comprehension twice. First dictionary: all numbers 1-20. 
Second dictionary: only even keys. Expected output: Two dictionaries printed.
"""

# 1. First dictionary: keys are numbers 1 to 20, values are their squares
squares_dict = {num: num**2 for num in range(1,21)}

# 2. Second dictionary: filter the first one to keep only even keys
even_squares_dict = {key: value for key, value in squares_dict.items() if key % 2 == 0}

# 3. Print the results
print(f"All squares (1-20): {squares_dict}\n")
print(f"Even squares only: {even_squares_dict}")
```

```python
"""Zadanie 12 – Analiza transakcji - słowniki zagnieżdżone
Masz listę transakcji (krotek): [('2024-01-15', 'Elektronika', 1200), ('2024-01-
16', 'Meble', 800), ('2024-01-15', 'Elektronika', 350), ...] . Stwórz słownik
gdzie kluczem jest data, a wartością słownik z sumą sprzedaży dla każdej kategorii tego
dnia.
Dataset:
transakcje = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]
Wymagania:
Zagnieżdżone słowniki: {data: {kategoria: suma}}
Iteruj po transakcjach i akumuluj sumy
Oczekiwany wynik:
Słownik typu: {'2024-01-15': {'Elektronika': 1200, 'Meble': 450, ...}, ...}

Task 12 – Transaction analysis - nested dictionaries
You have a list of transactions (tuples): [('2024-01-15', 'Elektronika', 1200), ...]. 
Create a dictionary where the key is the date, and the value is a dictionary with the 
total sales for each category on that day.
Requirements: Nested dictionaries {date: {category: sum}}. Iterate through transactions and accumulate sums.
"""

# 1. List of transactions (each transaction is a tuple: date, category, amount)
transacions = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]

# 2. Initialize an empty dictionary for our daily summaries
daily_sales = {}

#3 Iterate through the list and unpack each tuple into three variables
for date, category, amount in transacions:
    # Check if the DATE exists in our main dictionary
    if date not in daily_sales:
        # If not, create a new "drawer" for this date containing an empty dictionary
        daily_sales[date] = {}
    
    # Check if the CATEGORY exists in the dictionary for this specific date
    if category not in daily_sales[date]:
        #If not, create a "compartment" for this category and set its starting sum to 0
        daily_sales[date][category] = 0
    
    # Add the transaction amount to the correct date and category
    daily_sales[date][category] += amount

# 4. Print the final result:
print(f"Daily sales summary: {daily_sales}")     
```
