```python
"""
Zadanie 1 – Klasyfikacja liczb parzystych i nieparzystych
Utwórz tablicę NumPy zawierającą liczby od 1 do 100. Oblicz i wypisz:
Sumę wszystkich elementów
Średnią arytmetyczną
Medianę
Wartość maksymalną i minimalną
Wymagania:
Użyj np.arange() do stworzenia tablicy
Użyj funkcji agregujących NumPy ( np.sum , np.mean , etc.)
"""

import numpy as np

tworzenie_tablicy = np.arange(1,101)
suma_wszystkich_elementow =np.sum(tworzenie_tablicy)
srednia_arytmetyczna = np.mean(tworzenie_tablicy)
mediana = np.median(tworzenie_tablicy)
wartosc_maksymalna = np.max(tworzenie_tablicy)
wartosc_minimalna = np.min(tworzenie_tablicy)


print(f"Tablica Numpy: {tworzenie_tablicy}")
print(f"Suma wszystkich elementow: {suma_wszystkich_elementow}")
print(f"Średnia arytmetyczna: {srednia_arytmetyczna}")
print(f"Mediana: {mediana}")
print(f"Wartość maksymalna: {wartosc_maksymalna}")
print(f"Wartość minimalna: {wartosc_minimalna}")
```

```python
"""Zadanie 2 – Operacje wektorowe
Masz dwie tablice reprezentujące ceny produktów przed i po rabacie:

before = np.array([100, 250, 75, 320, 180, 450, 90])
after = np.array([80, 200, 60, 280, 150, 400, 75])

Oblicz i wypisz:
Różnicę między cenami (oszczędności)
Procent rabatu dla każdego produktu
Średnią oszczędność

Wymagania:
Użyj operacji wektorowych (bez pętli)
Wynik zaokrąglij do 2 miejsc po przecinku
"""

import numpy as np

before = np.array([100, 250, 75, 320, 180, 450, 90])
after = np.array([80, 200, 60, 280, 150, 400, 75])

obnizka = before - after
procent_rabatu = np.round((obnizka / before) * 100, 2)
srednia_oszczednosc = np.round(np.mean(obnizka), 2)

print(f"Róznica między cenami: {obnizka}")
print(f"\n Procent rabatu: {procent_rabatu}")
print(f"\n Średnia oszczędność: {srednia_oszczednosc}zł")
```

```python
"""Zadanie 3 – Indeksowanie i slicing
Dana tablica 1D:
temps = np.array([15, 18, 22, 25, 23, 20, 17, 14, 16, 19, 21, 24])
Wykonaj:
Wypisz temperatury z dni parzystych (0, 2, 4, ...)
Wypisz ostatnie 5 temperatur
Wypisz temperatury w odwrotnej kolejności
Znajdź temperatury większe niż 20
"""

import numpy as np

temps = np.array([15, 18, 22, 25, 23, 20, 17, 14, 16, 19, 21, 24])

print("Temperatury z dni parzystych:", temps[::2])
print("Ostatnie 5 temperatur:", temps[-5:])
print("Temperatury w odwrotnej kolejności:", temps[::-1])
print("Temperatury większe niż 20:", temps[temps>20])
```

```python
"""Zadanie 5 – Statystyki per wiersz i kolumna
Dana tablica reprezentująca sprzedaż 4 produktów w 7 dni:
sales = np.random.randint(10, 100, size=(4, 7))
Oblicz i wypisz:
Średnią sprzedaż per produkt (axis=1)
Średnią sprzedaż per dzień (axis=0)
Dzień z najwyższą sprzedażą (np.argmax)
Produkt z najniższą średnią sprzedażą
"""
import numpy as np

np.random.seed(42)
sales = np.random.randint(10, 100, size=(4, 7))

# 1. Średnie
srednia_per_produkt = np.round(np.mean(sales, axis=1),2)
srednia_per_dzien = np.round(np.mean(sales, axis=0),2)
# 2. Szukanie maksimów i minimów w wyliczonych już średnich
dzien_najnizsza_sprzedaz = np.argmax(srednia_per_dzien)
produkt_najnizsza_sprzedaz = np.argmax(srednia_per_produkt)

print("Sprzedaż 4 produktów w 7 dni:")
print(sales)
print("\n Średnią sprzedaż per produkt:", srednia_per_produkt)
print("Średnią sprzedaż per dzień", srednia_per_dzien)
print(f"\nDzień z najwyższą sprzedażą to dzien nr: {dzien_najnizsza_sprzedaz + 1}")
print(f"Produkt z najniższą średnią sprzedażą to produkt nr: {produkt_najnizsza_sprzedaz + 1}")
```

```python
"""Zadanie 6 – Boolean indexing
Masz tablicę z oceną studentów:
grades = np.array([78, 45, 92, 67, 34, 88, 56, 91, 23, 85, 73, 95])
Wykonaj:
Znajdź wszystkie oceny >= 80 (oceny wysokie)
Znajdź wszystkie oceny < 50 (oceny niedostateczne)
Policz ile jest ocen w przedziale [60, 80)
Zmień wszystkie oceny < 50 na 50 (podniesienie do minimum
"""

import numpy as np

grades = np.array([78, 45, 92, 67, 34, 88, 56, 91, 23, 85, 73, 95])

oceny_w_przedziale = np.sum((60 <= grades) & (grades < 80))

grades_copy = grades.copy()
grades_copy[grades_copy < 50] = 50

print("Wysokie wyniki:", grades[grades >= 80])
print("Oceny niedostateczne:", grades[grades < 50])
print("ilość ocen w przedziale [60, 80):", oceny_w_przedziale)
print("Podniesienie minimuoceny do minimum:", grades_copy)
```

```python
"""Zadanie 10 – Broadcasting w praktyce
Dana tablica ocen (5 studentów x 4 egzaminy):
exams = np.array([
    [78, 85, 92, 88],
    [65, 72, 68, 70],
    [90, 88, 95, 92],
    [55, 60, 58, 62],
    [82, 79, 85, 80]
])
Dodaj bonus do każdego egzaminu: [5, 10, 5, 10]
Następnie pomnóż wyniki każdego studenta przez mnożnik: [1.0, 1.1, 0.95, 1.15, 1.0]
Wymagania:
Użyj broadcastingu (bez pętli)
Wypisz wyniki przed i po
"""

import numpy as np

exams = np.array([
    [78, 85, 92, 88],
    [65, 72, 68, 70],
    [90, 88, 95, 92],
    [55, 60, 58, 62],
    [82, 79, 85, 80]
])

bonus = np.array([5, 10, 5, 10])
mnoznik = np.array([1.0, 1.1, 0.95, 1.15, 1.0])

exams_po_bonusie = exams + bonus 
mnoznik_pionowy = mnoznik.reshape(5, 1)
wyniki_ostateczne = exams_po_bonusie * mnoznik_pionowy

print("Oceny przed modyfikacją:")
print(exams)

print("\nOceny ostateczne:")
print(wyniki_ostateczne)
```

```python
"""Zadanie 11 – Analiza datasetu sprzedaży
Dataset: Sprzedaż 6 produktów w 30 dni
sales = np.random.randint(50, 200, size=(6, 30))
Oblicz:
Trend sprzedaży per produkt (różnica między ostatnimi 7 dniami a pierwszymi 7 dniami)
Najlepszy tydzień (dni 0-6, 7-13, 14-20, 21-27) dla całej sprzedaży
Wykres: sprzedaż per produkt w czasie (linie)
Wymagania:
Użyj slicingu do wycinania tygodni
Wizualizacja matplotlib
"""

import numpy as np
import matplotlib.pyplot as plt

sales = np.random.randint(50, 200, size=(6, 30))
#6 wierszy (axis=0) produkty, 30 dni wiersze (axis=1)

# CZĘŚĆ 1: Trend sprzedaży:
pierwsze_7_dni = np.sum(sales[:, :7], axis=1)
ostatnie_7_dni = np.sum(sales[:, -7:], axis=1)
trend = ostatnie_7_dni - pierwsze_7_dni

print("Trend (ostatnie 7 dni minus pierwsze 7 dni) dla 6 produktów:")
print(trend)

# CZĘŚĆ 2: Najlepszy tydzień:
tydzien_1 = np.sum(sales[:, 0:7])
tydzien_2 = np.sum(sales[:, 7:14])
tydzien_3 = np.sum(sales[:, 14-20])
tydzien_4 = np.sum(sales[:, 21:28])

lista_tygodni = [tydzien_1, tydzien_2, tydzien_3, tydzien_4]
najlepszy_indeks = np.argmax(lista_tygodni)
print(f"\nNajlepszy był tydzień nr {najlepszy_indeks + 1} z łączną sprzedażą: {lista_tygodni[najlepszy_indeks]}")

# CZĘŚĆ 3: Wykres liniowy
plt.figure(figsize=(10, 6))

for i in range(6):
    plt.plot(sales[i, :], label=f'Produkt {i+1}')

plt.title('Sprzedaż 6 produktów w ciągu 30 dni')
plt.xlabel('Dzień miesiąca')
plt.ylabel('Ilość sprzedanych sztuk')
plt.legend()
plt.grid(True)
plt.show()
```
